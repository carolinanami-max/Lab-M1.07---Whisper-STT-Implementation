```python
# STEP 1: TELL PYTHON WHERE TO GET DATA
print("ðŸ“¥ Going to download Titanic passenger list...")

# This is the link to the data
data_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

print(f"We'll use this link: {data_url}")
print("Ready to download!")
```

```python
# STEP 2: ACTUALLY DOWNLOAD THE DATA
import pandas as pd  # pd is a special tool for reading tables

print("ðŸ“Š Now downloading...")

# This line does the actual downloading
titanic_data = pd.read_csv(data_url)

print("âœ… SUCCESS! Data downloaded!")
print(f"We have information about {len(titanic_data)} passengers")
```

```python
# STEP 3: WHAT'S IN THIS SPREADSHEET?
print("ðŸ” Let's see what we have...")

# Show first 3 passengers
print("\nFIRST 3 PASSENGERS:")
print(titanic_data.head(3))

# Show column names
print("\nCOLUMNS (information we have for each passenger):")
for i, col in enumerate(titanic_data.columns, 1):
    print(f"{i}. {col}")

print(f"\nðŸ“ Shape: {titanic_data.shape[0]} rows Ã— {titanic_data.shape[1]} columns")
```

```python
# STEP 4: CHECK FOR BLANK SPOTS
print("ðŸ”Ž Checking for missing/blank information...")

# Count blanks in each column
blank_counts = titanic_data.isnull().sum()

print("\nBLANK SPOTS FOUND:")
for column_name, blank_count in blank_counts.items():
    if blank_count > 0:  # Only show columns with blanks
        percent_blank = (blank_count / len(titanic_data)) * 100
        print(f"  {column_name}: {blank_count} blanks ({percent_blank:.1f}%)")
```

```python
# STEP 5: DO SOME MATH
print("ðŸ“Š TIME FOR MATH!")
print("="*40)

# Let's pick the number columns
number_columns = ['Age', 'Fare', 'SibSp', 'Parch']

for col in number_columns:
    if col in titanic_data.columns:
        # Calculate statistics
        average = titanic_data[col].mean()
        middle_value = titanic_data[col].median()
        spread = titanic_data[col].std()

        print(f"\n{col.upper()}:")
        print(f"  Average: {average:.2f}")
        print(f"  Middle value: {middle_value:.2f}")
        print(f"  How spread out: {spread:.2f}")
```

```python
# STEP 6: CREATE NEW INFORMATION - FAMILY SIZE
print("âœ¨ CREATING NEW COLUMN: FAMILY SIZE")

# Make a copy so we don't mess up original
titanic_copy = titanic_data.copy()

# Family Size = Siblings + Parents + Yourself
titanic_copy['FamilySize'] = titanic_copy['SibSp'] + titanic_copy['Parch'] + 1

print("\nEXAMPLE (first 5 passengers):")
print(titanic_copy[['SibSp', 'Parch', 'FamilySize']].head())

print(f"\nAverage family size: {titanic_copy['FamilySize'].mean():.2f}")
```

```python
# STEP 7: CREATE "IS ALONE"
print("âœ¨ CREATING NEW COLUMN: IS ALONE?")

# Is Alone = True if FamilySize is 1
titanic_copy['IsAlone'] = titanic_copy['FamilySize'] == 1

print("\nEXAMPLE (first 5 passengers):")
print(titanic_copy[['FamilySize', 'IsAlone']].head())

alone_percent = titanic_copy['IsAlone'].mean() * 100
print(f"\nPercentage traveling alone: {alone_percent:.1f}%")
```

```python
# STEP 8: CREATE AGE GROUPS
print("âœ¨ CREATING NEW COLUMN: AGE GROUP")

def put_in_age_group(age):
    """Put each age into a category"""
    if pd.isna(age):  # If age is blank
        return 'Unknown'
    elif age < 18:
        return 'Child'
    elif age < 30:
        return 'Young Adult'
    elif age < 50:
        return 'Adult'
    else:
        return 'Senior'

# Apply the function to every age
titanic_copy['AgeGroup'] = titanic_copy['Age'].apply(put_in_age_group)

print("\nEXAMPLE (first 5 passengers):")
print(titanic_copy[['Age', 'AgeGroup']].head())

print("\nCOUNT IN EACH GROUP:")
print(titanic_copy['AgeGroup'].value_counts())
```

```python
# STEP 9: DO NEW FEATURES PREDICT SURVIVAL?
print("ðŸ”¬ DO OUR NEW FEATURES HELP?")
print("="*50)

# Split into two groups
survivors = titanic_copy[titanic_copy['Survived'] == 1]
not_survivors = titanic_copy[titanic_copy['Survived'] == 0]

print("\n1. FAMILY SIZE:")
print(f"  Survivors' average: {survivors['FamilySize'].mean():.2f}")
print(f"  Non-survivors' average: {not_survivors['FamilySize'].mean():.2f}")
difference = abs(survivors['FamilySize'].mean() - not_survivors['FamilySize'].mean())
print(f"  Difference: {difference:.2f}")

print("\n2. ALONE vs WITH FAMILY:")
print(f"  % of survivors alone: {survivors['IsAlone'].mean()*100:.1f}%")
print(f"  % of non-survivors alone: {not_survivors['IsAlone'].mean()*100:.1f}%")

print("\n3. AGE GROUPS SURVIVAL RATES:")
for group in ['Child', 'Young Adult', 'Adult', 'Senior', 'Unknown']:
    group_data = titanic_copy[titanic_copy['AgeGroup'] == group]
    if len(group_data) > 0:
        survival_rate = group_data['Survived'].mean() * 100
        print(f"  {group}: {survival_rate:.1f}% survived")
```

```python
# STEP 10: PREPARE FOR JSON (SAVING)
import json # Import the json module

print("ðŸ’¾ PREPARING TO SAVE...")

# Create a list to hold all passenger dictionaries
all_passengers = []

# Convert first 10 passengers (for demo)
for i in range(min(10, len(titanic_copy))):
    passenger = titanic_copy.iloc[i]

    passenger_dict = {
        'id': int(passenger['PassengerId']),
        'name': str(passenger['Name']),
        'age': float(passenger['Age']) if pd.notna(passenger['Age']) else None,
        'sex': str(passenger['Sex']),
        'survived': bool(passenger['Survived']),
        'pclass': int(passenger['Pclass']),
        'family_size': int(passenger['FamilySize']),
        'is_alone': bool(passenger['IsAlone']),
        'age_group': str(passenger['AgeGroup'])
    }

    all_passengers.append(passenger_dict)

print(f"Created dictionaries for {len(all_passengers)} passengers")
print("\nExample of one passenger's data:")
print(json.dumps(all_passengers[0], indent=2))
```

```python
# STEP 11: SAVE TO JSON FILE
print("ðŸ’¾ SAVING TO FILE...")

from datetime import datetime # Import datetime to get current time

# Create the complete data structure
final_data = {
    'metadata': {
        'title': 'Titanic Passenger Analysis',
        'total_passengers_analyzed': len(all_passengers),
        'created_date': str(datetime.now())
    },
    'statistics': {
        'average_age': float(titanic_copy['Age'].mean()),
        'average_fare': float(titanic_copy['Fare'].mean()),
        'survival_rate': float(titanic_copy['Survived'].mean() * 100)
    },
    'passengers': all_passengers
}

# Save to file
with open('titanic_analysis.json', 'w') as f:
    json.dump(final_data, f, indent=2)

print("âœ… SAVED to 'titanic_analysis.json'")

# Download it from Colab
from google.colab import files
files.download('titanic_analysis.json')
print("ðŸ“¥ File downloading to your computer now!")
```
