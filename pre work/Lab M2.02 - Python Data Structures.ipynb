```python
# Exercise 1: Working with Lists
# Creating a guest list for an event

# Step 1: Create a list with at least 5 guest names
guest_list = ["Alice", "Bob", "Charlie", "Diana", "Eve"]

print("Original guest list:")
print(guest_list)
print()

# Step 2: Add two more guests to the end of the list using .append()
guest_list.append("Frank")
guest_list.append("Grace")

print("After adding two more guests:")
print(guest_list)
print()

# Step 3: Print the total number of guests using len()
print(f"Total number of guests: {len(guest_list)}")
print()

# Step 4: Print the first guest and the last guest using indexing
print(f"First guest: {guest_list[0]}")
print(f"Last guest: {guest_list[-1]}")
print()

# Step 5: Sort the list alphabetically using .sort()
guest_list.sort()

# Step 6: Print the sorted list
print("Sorted guest list (alphabetically):")
print(guest_list)
```

```python
# Exercise 2: Working with Dictionaries
# Storing contact information for guests

# Step 1: Create a dictionary with at least 3 key-value pairs
contact_info = {
    "Alice": "alice@email.com",
    "Bob": "bob@email.com", 
    "Charlie": "charlie@email.com"
}

print("Initial contact information:")
print(contact_info)
print()

# Step 2: Add 2 more guests and their emails to the dictionary
contact_info["Diana"] = "diana@email.com"
contact_info["Eve"] = "eve@email.com"

print("After adding two more guests:")
print(contact_info)
print()

# Step 3: Print the email address of one specific guest by using their name as the key
print(f"Alice's email address: {contact_info['Alice']}")
print(f"Bob's email address: {contact_info['Bob']}")
print()

# Step 4: Print all the keys (guest names) using .keys()
print("All guest names:")
print(contact_info.keys())
print()

# Step 5: Print all the values (email addresses) using .values()
print("All email addresses:")
print(contact_info.values())
print()

# Step 6: Check if a specific guest exists in the dictionary using the in keyword
# Check for existing guest
guest_to_check = "Charlie"
if guest_to_check in contact_info:
    print(f"{guest_to_check} exists in the contact list. Email: {contact_info[guest_to_check]}")
else:
    print(f"{guest_to_check} does not exist in the contact list.")
print()

# Check for non-existing guest
guest_to_check = "Frank"
if guest_to_check in contact_info:
    print(f"{guest_to_check} exists in the contact list. Email: {contact_info[guest_to_check]}")
else:
    print(f"{guest_to_check} does not exist in the contact list.")
print()

# Additional demonstration: Show all key-value pairs
print("Complete contact list (key-value pairs):")
for name, email in contact_info.items():
    print(f"{name}: {email}")
```

```python
# Exercise 3: Working with Sets
# Collecting unique interests from guests

# Step 1: Create a set called interests with at least 5 different interests
interests = {"music", "reading", "sports", "cooking", "travel"}

print("Initial set of interests:")
print(interests)
print(f"Number of interests: {len(interests)}")
print()

# Step 2: Add 2 more interests using .add()
interests.add("photography")
interests.add("gardening")

print("After adding two more interests:")
print(interests)
print(f"Number of interests: {len(interests)}")
print()

# Step 3: Try adding a duplicate interest and observe what happens
print("Trying to add duplicate interest 'music':")
interests.add("music")  # This won't add anything since "music" already exists
print("Set after trying to add 'music' again:")
print(interests)
print(f"Number of interests (still the same): {len(interests)}")
print()

# Step 4: Print the set and notice that it doesn't maintain order
print("Sets do not maintain order (may appear different each time):")
print("Notice that the order of elements in a set is arbitrary")
print("Current order:", interests)
print()

# Let's demonstrate by creating the same set multiple times
print("Demonstrating unordered nature of sets:")
set1 = {"a", "b", "c", "d"}
set2 = {"d", "c", "b", "a"}
print(f"set1 (created in 'a', 'b', 'c', 'd' order): {set1}")
print(f"set2 (created in 'd', 'c', 'b', 'a' order): {set2}")
print(f"Are they equal? {set1 == set2}")
print()

# Step 5: Create a second set with some overlapping interests
second_interests = {"cooking", "travel", "hiking", "yoga", "movies"}

print("Second set of interests:")
print(second_interests)
print()

# Step 6: Find the common interests between the two sets
# Using .intersection() method
common_interests = interests.intersection(second_interests)
print("Common interests (using .intersection()):")
print(common_interests)
print()

# Using the & operator (alternative way)
common_interests_operator = interests & second_interests
print("Common interests (using & operator):")
print(common_interests_operator)
print()

# Show what's unique to each set
unique_to_first = interests - second_interests
unique_to_second = second_interests - interests

print("Interests unique to first set:")
print(unique_to_first)
print()

print("Interests unique to second set:")
print(unique_to_second)
print()

# Additional demonstration: Show all interests combined (union)
all_interests = interests.union(second_interests)
print("All unique interests combined (union):")
print(all_interests)
print(f"Total unique interests: {len(all_interests)}")
```

```python
# Comparison Reflection

## When would you use a list?
I would use a **list** when:
- I need to maintain a specific order of elements (like a sequence or queue)
- I need to store duplicate values
- I need to access elements by their position/index
- I'm working with a simple ordered collection of items

Lists are special because they preserve insertion order, allow duplicates, and provide index-based access. They're perfect for tasks like maintaining guest lists, storing data in sequence, or any collection where position matters.

## When would you use a dictionary?
I would use a **dictionary** when:
- I need to associate keys with values (like a phonebook or database)
- I need fast lookups based on a unique identifier
- I'm working with labeled data where each piece has a specific name/key
- I need to store structured data with named attributes

Dictionaries are special because they provide O(1) average time complexity for lookups, use key-value pairs for organized data storage, and ensure keys are unique. They're ideal for storing contact information, configuration settings, or any data that needs to be retrieved by name rather than position.

## When would you use a set?
I would use a **set** when:
- I need to ensure all elements are unique (removing duplicates)
- I need to perform set operations like union, intersection, or difference
- I need fast membership testing (checking if an item exists)
- The order of elements doesn't matter
- I'm working with mathematical sets or collections of unique items

Sets are special because they automatically eliminate duplicates, provide extremely fast membership testing (O(1) average), and support mathematical set operations. They're perfect for finding common interests, removing duplicate entries, or tracking unique visitors.

## What's the main difference between a list and a set?
The main differences are:

1. **Uniqueness**: Lists allow duplicate elements, while sets only store unique elements
2. **Order**: Lists maintain insertion order, while sets are unordered
3. **Indexing**: Lists support index-based access (e.g., `list[0]`), while sets don't
4. **Performance**: Sets provide faster membership testing (O(1) average) compared to lists (O(n))
5. **Mutability**: Both are mutable, but sets require elements to be hashable

## Can you access dictionary values by index like you do with lists? Why or why not?

**No**, you cannot access dictionary values by numerical index like you do with lists. Here's why:

1. **Key-based access**: Dictionaries use keys (not positions) to access values
   - ✅ Correct: `dict["key_name"]`
   - ❌ Incorrect: `dict[0]` (unless "0" is actually a string key)

2. **Unordered nature**: In Python versions before 3.7, dictionaries didn't maintain order, so index-based access wouldn't make sense. While Python 3.7+ dictionaries preserve insertion order, they're still fundamentally key-based data structures.

3. **Alternative approaches**: If you need index-like access, you can:
   - Convert keys to a list and access by position: `list(dict.keys())[0]`
   - Use `.values()` or `.items()` with list conversion
   - Consider using `collections.OrderedDict` for more control

4. **Philosophical difference**: Lists are for ordered sequences; dictionaries are for key-value mappings. Using index-based access on a dictionary would confuse the purpose of these data structures.

The proper way to access dictionary values is always through their keys, which makes the intent clear and maintains the semantic meaning of the data structure.
```
